```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = 1.5
y = 0.5
```

```python
w = np.random.random()
print(w)
```

```python
alpha =0.1
epochs = 50
wlist = []
elist = []
for epoch in range(epochs):
    #forward pass
    y_pred = x*w
    #error
    error = np.square(y-y_pred)
    
    #compute gradient
    gradient = -2 * (y-y_pred)*x
    
    #back propagation
    w = w-alpha*gradient
    wlist.append(w)
    elist.append(error)
```

```python
plt.figure(figsize=(8,8))
plt.plot(range(epochs),elist)
plt.xlabel("epochs")
plt.show()
```
